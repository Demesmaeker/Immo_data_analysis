# Price / House area

```python
price_house_area = clean(df)
```

```python
price_house_area
```

```python
def get_house_area(house_area):
    if house_area < 500:
        return "Below 500"
    elif house_area < 1000:
        return "500-1000"
    elif house_area < 1500:
        return "1000-1500"
    elif house_area < 2000:
        return "1500-2000"
    elif house_area < 2500:
        return "2000-2500"
    else:
        return "Over 2500"
```

```python
price_house_area['house_area_grouped'] = price_house_area['house_area'].apply(get_house_area)
```

```python
price_house_area
```

```python
price_house_area = price_house_area.groupby("house_area_grouped").price.agg(["mean"])
```

```python
price_house_area
```

```python
price_house_area.reset_index(level=0, inplace=True)
```

```python
price_house_area.plot.bar(x = "house_area_grouped", y = "mean")
```

# Price / Garden area

```python
def get_garden_area(garden_area):
    if garden_area == 0:
        return "No garden"
    elif garden_area < 20:
        return "Below 20"
    elif garden_area <= 50:
        return "20-50"
    elif garden_area <= 100:
        return "50-100"
    elif garden_area <= 200:
        return "100-200"
    elif garden_area <= 300:
        return "200-300"
    elif garden_area <= 400:
        return "300-400"
    else:
        return "Over 500"
```

```python
price_garden_area['garden_area_grouped'] = price_garden_area['garden_area'].apply(get_garden_area)
```

```python
price_garden_area
```

```python
price_garden_area = price_garden_area.groupby("garden_area_grouped").price.agg(["mean"])
```

```python
price_garden_area
```

```python
price_garden_area.reset_index(inplace=True)
```

```python
price_garden_area
```
