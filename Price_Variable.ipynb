# Price / Locality

```python
price_locality = clean(df)
```

```python
price_locality
```

```python
province_price = price_locality.groupby("province").price.agg(["mean"])
```

```python
province_price
```

```python
province_price.reset_index(level=0, inplace=True)
```

```python
province_price.plot.bar(x = "province", y = "mean")
```
